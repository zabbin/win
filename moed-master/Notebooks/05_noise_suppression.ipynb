```python
import sys
from random import random
sys.path.append('../')
```

```python
from moed.model import Sequence, Model
from moed.analysis import Analysis
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
_range = [x / 100 for x in range(1000)]

signal = Model.harmonic(_range)
signal.plot()
```

```python
N = 10
signals = []
for x in range(N):
    noise = Sequence.from_func(_range, lambda x: random()*2)
    signals.append(signal + noise)

signals[0].plot()

res = Analysis.average(signals)
res.plot()
```
