```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
from moed.model import Model
from moed.io import IO
```

```python
x = range(1, 1001, 1)
a = 0.5
b = 3
alpha = 0.005
beta = 1

y1 = Model.linear(x, a, b)

y2 = Model.linear(x, -a, b)

y3 = Model.exponential(x, alpha, beta)

y4 = Model.exponential(x, -alpha, beta)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes[0, 0].plot(x, y1)
axes[0, 1].plot(x, y2)
axes[1, 0].plot(x, y3)
axes[1, 1].plot(x, y4)
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.plot(x[:500], y1[:500])
plt.plot(x[500:750], y2[500:750])
plt.plot(x[750:1001], y3[750:1001])
plt.show()
```

```python
x = IO.normalize(x)
y1 = IO.normalize(y1)
y1 = IO.normalize(y2)
y1 = IO.normalize(y3)
y1 = IO.normalize(y4)

fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes[0, 0].plot(x, y1)
axes[0, 1].plot(x, y2)
axes[1, 0].plot(x, y3)
axes[1, 1].plot(x, y4)
plt.show()
```
