```python
import sys
from random import random
sys.path.append('../')

from moed.model import Sequence, Model
from moed.analysis import Analysis
from moed.proc import Proc

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
_range = range(1000)
```

```python
noise = Sequence.from_func(_range, lambda x: random())
my_noise = Sequence.from_func(_range, lambda x: Proc.my_random())

noise.plot()
my_noise.plot()
```

```python
autocorr = Analysis.auto_correlation(noise)
my_autocorr = Analysis.auto_correlation(my_noise)
cross = Analysis.cross_correlation(noise, my_noise)

autocorr.plot()
my_autocorr.plot()
cross.plot()
```

```python
g = Analysis.gist(noise, 5)
mg = Analysis.gist(my_noise, 5)

g.plot()
mg.plot()
```

```python
avg = Analysis.avg(noise)
my_avg = Analysis.avg(my_noise)
print(f"Average\nBuiltIn:\t{avg:.3f}\nCustom: \t{my_avg:.3f}")
```

```python
var = Analysis.var(noise)
my_var = Analysis.var(my_noise)
print(f"Variance\nBuiltIn:\t{var:.3f}\nCustom: \t{my_var:.3f}")
```

```python
skewness = Analysis.skewness(noise)
my_skewness = Analysis.skewness(my_noise)
print(f"Skewness\nBuiltIn:\t{skewness:.3f}\nCustom: \t{my_skewness:.3f}")
```

```python
excess = Analysis.excess(noise)
my_excess = Analysis.excess(my_noise)
print(f"Excess\nBuiltIn:\t{excess:.3f}\nCustom: \t{my_excess:.3f}")
```

```python
MS = Analysis.MS(noise)
my_MS = Analysis.MS(my_noise)
print(f"MS\nBuiltIn:\t{MS:.3f}\nCustom: \t{my_MS:.3f}")
```

```python
MSE = Analysis.MSE(noise)
my_MSE = Analysis.MSE(my_noise)
print(f"MSE\nBuiltIn:\t{MSE:.3f}\nCustom: \t{my_MSE:.3f}")
```

```python
assymmetry = Analysis.assymmetry(noise)
my_assymmetry = Analysis.assymmetry(my_noise)
print(f"Assymmetry\nBuiltIn:\t{assymmetry:.3f}\nCustom: \t{my_assymmetry:.3f}")
```

```python
kurtosis = Analysis.kurtosis(noise)
my_kurtosis = Analysis.kurtosis(my_noise)
print(f"Kurtosis\nBuiltIn:\t{kurtosis:.3f}\nCustom: \t{my_kurtosis:.3f}")
```

```python
mini = min(noise.y)
my_mini = min(my_noise.y)
print(f"MIN\nBuiltIn:\t{mini:.3f}\nCustom: \t{my_mini:.3f}")
```

```python
maxi = max(noise.y)
my_maxi = max(my_noise.y)
print(f"MAX\nBuiltIn:\t{maxi:.3f}\nCustom: \t{my_maxi:.3f}")
```
