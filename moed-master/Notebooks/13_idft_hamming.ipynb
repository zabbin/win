```python
_range = [x / 10 for x in range(1000)]

signal = Model.harmonic(_range)
signal.plot()
```

```python
dft = Analysis.dft_complex(signal)
idft = Analysis.idft(dft).plot()
```

```python
dft = Analysis.dtf(signal)
dft.plot()
```

```python
_range1 = [x / 10 for x in range(1000)]
_range2 = [x / 10 for x in range(100, 900)]
signal = Sequence.from_func(_range1, lambda x: 0) + Model.harmonic(_range2)
signal.plot()
Analysis.dtf(signal).plot()
```

```python
ham = signal.hamming()
dft = Analysis.dtf(ham)
ham.plot()
dft.plot()
```
