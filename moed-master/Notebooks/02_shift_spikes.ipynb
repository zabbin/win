```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from moed.model import Model
from moed.io import IO
from moed.proc import Proc
```

## Shift

```python
x = range(1, 1001, 1)
a = 0.5
b = 3
alpha = 0.005
beta = 1

s= 10
N= 1000

y1= np.random.uniform (low=-s, high=s, size=N)
y2 = Proc.shift(y1, 400, 600, 50)
```

```python
fig, axes = plt.subplots(2, figsize=(15, 10))
axes[0].plot(x, y1)
axes[1].plot(x, y2)
plt.show()
```

## Random

```python
y1= np.random.uniform (low=-s, high=s, size=N)
y2 = [Proc.my_random(end=2*s)-s for x in range(1000)]
```

```python
fig, axes = plt.subplots(2, figsize=(15, 10))
axes[0].plot(x, y1)
axes[1].plot(x, y2)
plt.show()
```

## Spikes

```python
y1= np.random.uniform (low=-s, high=s, size=N)
y2 = Proc.spikes(y1, 20, 3)
```

```python
fig, axes = plt.subplots(2, figsize=(15, 10))
axes[0].plot(x, y1)
axes[1].plot(x, y2)
plt.show()
```
