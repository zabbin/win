```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from moed.model import Model
from moed.io import IO
from moed.proc import Proc
```

```python
s=10
N=1000
x=range(N)

bi_rand = np.random.uniform (low=-s, high=s, size=N)
my_rand = [Proc.my_random(end=2*s)-s for x in range(N)]
fig, axes = plt.subplots(2, figsize=(15, 10))
axes[0].plot(x, bi_rand)
axes[1].plot(x, my_rand)
plt.show()
```

```python
from statistics import mean, stdev
```

```python
interval = 100
bi_mean = [mean(bi_rand[i:i+interval]) for i in range(N-interval)]
my_mean = [mean(my_rand[i:i+interval]) for i in range(N-interval)]
bi_stdev = [stdev(bi_rand[i:i+interval]) for i in range(N-interval)]
my_stdev = [stdev(my_rand[i:i+interval]) for i in range(N-interval)]
```

```python
fig, axes = plt.subplots(2, figsize=(15, 10))
axes[0].plot(x, bi_rand)
axes[0].plot(x[50:-50], bi_mean)
axes[0].plot(x[50:-50], bi_stdev)
axes[1].plot(x, my_rand)
axes[1].plot(x[50:-50], my_mean)
axes[1].plot(x[50:-50], my_stdev)
plt.show()
```
